```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('creditcard.csv')

# Displaying basic information about the dataset
print(data.info())

# Displaying the first few rows of the dataset
print(data.head())

# Checking for missing values
print("Missing values:", data.isnull().sum().max())

# Exploring the distribution of the target variable 'Class'
class_distribution = data['Class'].value_counts()
print("Class Distribution:\n", class_distribution)

```

```python
# Visualizing the class distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=data)
plt.title('Distribution of Classes (0: Not Fraud, 1: Fraud)')
plt.show()

```

```python
# Exploring statistical summary of transaction amounts for both classes
amount_summary = data.groupby('Class')['Amount'].describe()
print("Transaction Amount Summary:\n", amount_summary)

```

```python
# Visualizing transaction amounts for both classes
plt.figure(figsize=(6, 4))
sns.boxplot(x='Class', y='Amount', data=data)
plt.title('Transaction Amount Distribution by Class')
plt.show()

```

```python
# Exploring time distribution for both classes
plt.figure(figsize=(6, 4))
sns.distplot(data[data['Class'] == 0]['Time'], label='Not Fraud', hist=False)
sns.distplot(data[data['Class'] == 1]['Time'], label='Fraud', hist=False)
plt.title('Distribution of Transaction Time by Class')
plt.legend()
plt.show()

```

```python
# Correlation heatmap
plt.figure(figsize=(6,4))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, cmap="coolwarm", annot=False)
plt.title('Correlation Heatmap')
plt.show()
```

<b> Interpretation </b> 

This EDA covers the following aspects:

Dataset Overview:
Displaying basic information about the dataset, such as columns and data types.
Showing the first few rows to get a sense of the data.

Class Distribution:
Examining the distribution of the target variable 'Class' (0: Not Fraud, 1: Fraud) to understand the class imbalance.

Transaction Amount Analysis:
Investigating the statistical summary of transaction amounts for both classes.
Visualizing the distribution of transaction amounts using boxplots.

Transaction Time Analysis:
Exploring the distribution of transaction time for both fraud and non-fraud transactions.

Correlation Heatmap:
Generating a heatmap to visualize the correlation between different features in the dataset.
This EDA provides valuable insights into the distribution of classes, transaction amounts, and time. Understanding these patterns is crucial for building effective models for credit card fraud detection.
